<h2 style='font-size:30px'> 9.1 Parametric Equations</h2>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Em curvas que ferem o Teste da Linha Vertical, não podemos expressá-las como $y=f(x)$ (uma mesma abscissa mapeia mais de uma ordenada).
        </li>
        <li>
            Nessa situação, é possível que, na verdade, tenhamos que representar o valor de cada coordenada como uma função separada definida por uma terceira variável $t$.
        </li>
        <li>
            Em um problema de equações paramétricas, utilizamos os seguites termos:
            <ul> 
                <li>
                    <i> Parâmetro</i>: Seria a variável usada para definir o valor das nossas coordenadas. 
                </li>
                <li>
                    <i> Equação Paramétrica </i>: Funções baseadas pelo parâmetro que definirão o valor de nossas coordenadas.  
                </li>
                <li>
                    <i> Curva Paramétrica</i>: A curva gerada pelo resultado das equações paramétricas para cada valor de $t$. 
                </li>
            </ul>
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    Example I
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
        </li>
    </ul>
</div>

```python
from matplotlib.pyplot import plot
import numpy as np

t = np.linspace(-10, 10, 1001)
x = t**2 - 2*t
y = t+1

plot(x,y)
```

<div> 
    <ul style='font-size:20px'> 
        <li>
            Algumas vezes, vamos querer estabelecer um intervalo em que $t$ deve residir $a\leq{t}\leq{b}$.
        </li> 
        <li> 
            Nesse caso, os pontos $(f(a),g(a))$ e $(f(b),g(b))$ são conhecidos como <i> initial point</i> e <i> terminal point</i>, respectivamente.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    Example II
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
        </li>
    </ul>
</div>

```python
from matplotlib.pyplot import plot
import numpy as np

t = np.linspace(0, 2*np.pi ,101)
x = np.cos(t)
y = np.sin(t)

plot(x,y)
```

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    Example III
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
        </li>
    </ul>
</div>

```python
from matplotlib.pyplot import plot
import numpy as np

t = np.linspace(0, 2*np.pi ,101)
x = np.sin(2*t)
y = np.cos(2*t)

plot(x,y)
```

<h3 style='font-size:30px;font-style:italic'> Curve x Parametric Curve</h3>
<div> 
    <ul style='font-size:20px'>
        <li>
            Repare que, apesar de os exemplos II e III gerarem a mesma curva, o seu traçado ocorre de maneira distinta em cada um deles. Em II, o círculo é gerado no sentido anti-horário, enquanto que em III, no horário, passando sempre duas vezes pelos mesmos pontos.
        </li>
        <li>
            É por conta dessas diferenças de traçado que podem ocorrer que distinguimos uma <i>curva</i> de uma <i>curva paramétrica</i>. A primeira seria a figura final gerada pelas equações, enquanto que a segunda o traçado delineado por elas.
        </li>
     </ul>
</div>

```python
# Note que-1<=sin(t)<=1. Portanto, os valores de "x" sempre ficarão dentro desse intervalo, independentemente
# de "t".

# Podemos entender também y=x**2, justificando o formato de parábola da curva.
# No entanto, como sin é uma função periódica, a curva paramétrica das equações seria a curva da parábola 
# desenhada de ida e volta várias vezes.
from matplotlib.pyplot import plot
import numpy as np

t = np.linspace(-10, 10 ,1001)
x = np.sin(t)
y = np.sin(t) ** 2

plot(x,y)
```

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    Equação de um Ciclóide
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A demonstração da equação de um ciclóide pode ser realizada geometricamente.
        </li>
        <li>
            Dada uma circunferência apoiada sobre o eixo-x, cujo ciclóide é traçado por um ponto $P$, podemos expressar as coordenadas de seu centro como $C(r\theta,r)$. No caso, $\theta$ seria o ângulo formado entre a reta $\overline{CP}$ e a que liga $C$ com o eixo $x$ ($\overline{TC}$). 
        </li>
        <li>
            Podemos usar essas retas, junto com uma que ligue $P$ a $\overline{TC}$, para formar um triângulo retângulo, como mostra a imagem.
        </li>
    </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'> 
                     <img src='cycloid-geometry.png' >
        <figcaption style='font-size:15px'>
        </figcaption>
    </center>
</figure>

<div> 
    <ul style='font-size:20px'> 
        <li>
            A partir desse triângulo, podemos deduzir as coordenadas do ponto $P$ como as seguintes equações paramétricas:
            <ul> 
                <li>
                    $x$ = $r\theta-r\sin{\theta}=r(\theta-\sin{\theta})$
                </li>
                <li>
                    $y$ = $r-r\cos{\theta}=r(1-\cos{\theta})$
                </li>
            </ul>
        </li>
    </ul>
</div>
