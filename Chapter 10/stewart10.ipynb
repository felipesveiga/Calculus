<h2 style='font-size:30px'> 10.1 Parametric Equations</h2>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Em curvas que ferem o Teste da Linha Vertical, não podemos expressá-las como $y=f(x)$ (uma mesma abscissa mapeia mais de uma ordenada).
        </li>
        <li>
            Nessa situação, é possível que, na verdade, tenhamos que representar o valor de cada coordenada como uma função separada definida por uma terceira variável $t$.
        </li>
        <li>
            Em um problema de equações paramétricas, utilizamos os seguites termos:
            <ul> 
                <li>
                    <i> Parâmetro</i>: Seria a variável usada para definir o valor das nossas coordenadas. 
                </li>
                <li>
                    <i> Equação Paramétrica </i>: Funções baseadas pelo parâmetro que definirão o valor de nossas coordenadas.  
                </li>
                <li>
                    <i> Curva Paramétrica</i>: A curva gerada pelo resultado das equações paramétricas para cada valor de $t$. 
                </li>
            </ul>
        </li>
    </ul>
</div>

        </li>

        </li>
        <li>
            Podemos usar essas retas, junto com uma que ligue $P$ a $\overline{TC}$, para formar um triângulo retângulo, como mostra a imagem.

<figure>

</figure>

<div> 
    <ul style='font-size:20px'> 
        <li>
            A partir desse triângulo, podemos deduzir as coordenadas do ponto $P$ como as seguintes equações paramétricas:
            <ul> 
                <li>
                    $x$ = $r\theta-r\sin{\theta}=r(\theta-\sin{\theta})$
                </li>
                <li>
                    $y$ = $r-r\cos{\theta}=r(1-\cos{\theta})$
                </li>

<h2 style='font-size:30px'> 10.2 Calculus with Parametric Curves</h2>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Nessa seção, exploramos a aplicação de cálculo sobre as equações paramétricas.
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Tangents</h3>
<div> 
    <ul style='font-size:20px'>
        <li>
            Pelo fato de as coordenadas dos pontos se darem por uma variável latente ao espaço, computar a derivada de curvas paramétricas pode ser uma tarefa complexa.
        </li>
        <li>
            A fim de facilitar esse problema, podemos nos aproveitar de um fenômeno ocorrido em algumas equações. Dadas as equações paramétricas $x=f(t)$ e $y=g(t)$, conseguimos expressar $y=F(x)$.
        </li>
        <li>
            Uma vez que conseguimos encontrar uma relação $x\to{y}$, tornamos possível a modelagem de $\frac{dy}{dx}$, o que nos permite encontrar a derivada da curva em qualquer ponto!
        </li>
     </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'> 
        $g(t)=F(f(t))$
    </center>
</figure>

<div> 
    <ul style='font-size:20px'>
        <li>
            Supondo que $f$, $g$ e $F$ são funções diferenciáveis, podemos definir $g'(t)$ recorrendo à Chain Rule.
        </li>
     </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'> 
        $$
        \begin{align*}g'(t)&=F'(f(t))f'(t)=F'(x)f'(t)\\
        F'(x)&=\frac{g'(t)}{f'(t)}\text{, }f'(t)\neq{0}\\
        \frac{dy}{dx}&=\frac{\frac{dy}{dt}}{\frac{dx}{dt}}\text{, }\frac{dy}{dt}\neq{0}
        \end{align*}
        $$
    </center>
</figure>

<div> 
    <ul style='font-size:20px'>
        <li>
            Com isso, conseguimos mensurar a derivada da curva sem sequer precisarmos eliminar o parâmetro $t$! 
        </li>
        <li>
            Para qualquer ponto $(x,F(x))$, sua derivada sempre terá o valor $F'(x)$.            
        </li>
     </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> 
    $\frac{d^{2}y}{dx^{2}}$
</h4>
<div> 
    <ul style='font-size:20px'> 
        <li>
            A partir da definição, acima, podemos deduzir a segunda derivada de $F(x)$ inputando $\frac{dy}{dx}$ no lugar de $y$.
        </li>
    </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'> 
        $$
        \begin{align*}
        \frac{d^{2}y}{dx^{2}}=\frac{d}{dx}\left[\frac{dy}{dx}\right]=\frac{\frac{d}{dx}\frac{dy}{dt}}{\frac{dx}{dt}}=\frac{\frac{d}{dt}\frac{dy}{dx}}{\frac{dx}{dt}}
        \end{align*}
        $$
    </center>
</figure>

<h3 style='font-size:30px;font-style:italic'> Areas</h3>
<div> 
    <ul style='font-size:20px'>
        <li>
            Dadas as equações paramétricas $x=f(t)$ e $y=g(t)$, poderíamos querer mensurar a integral $\int_{a}^{b}{ydx}$.
        </li>
        <li>
            Para conseguirmos fazer isso, poderíamos aplicar algumas substituições que possibilitem atingirmos o resultado usando $f(t)$ e $g(t)$.
        </li>
     </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'> 
        $\displaystyle \int_{a}^{b}{ydx}=\int_{\alpha}^{\beta}{g(t)f'(t)dt}$
    </center>
</figure>

<h3 style='font-size:30px;font-style:italic'> Arc Length</h3>
<div> 
    <ul style='font-size:20px'>
        <li>
            Seja $C$ uma curva definida pelas equações paramétricas $x=f(t)$ e $y=g(t)$, poderíamos querer estimar o seu comprimento no intervalo de abscissas $[a,b]$.
        </li>
        <li>
            Uma maneira de fazermos isso seria definindo infinitos pontos sobre a curva $P(x_{i},y_{i})\text{, }x_{i}=f(t_{i})\text{, }y_{i}=g(t_{i})\text{, }t_{i}\in[\alpha,\beta]$, somando as distâncias entre pares.
        </li>
        <li>
            Considere que $a=f(\alpha)$ e $b=g(\beta)$.
        </li>
     </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'> 
        $$
        \displaystyle
        \begin{align*}
            L=\lim_{n\to\infty}\sum_{i=1}^{n}{\overline{P_{i-1}P_{i}}} \\
            \overline{P_{i-1}P_{i}}=\sqrt{(\Delta{x}_{i})^{2}+(\Delta{y}_{i})^{2}} 
        \end{align*}
        $$
        <figcaption style='font-size:15px'> 
            $\overline{P_{i-1}P_{i}}$ é a distância entre $P_{i-1}$ e $P_{i}$, podendo ser expressa como a hipotenusa do triângulo retângulo criado ao ligarmos esses pontos.
        </figcaption>
    </center>
</figure>

<div> 
    <ul style='font-size:20px'>
        <li>
            Sejam $f'$ e $g'$, as derivadas das equações, podemos expressar a inclinação da hipotenusa do $i$-ésimo intervalo como: 
        </li>
     </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'> 
        $$
        \displaystyle
        \begin{align*}
            f_{i}'=\frac{f(t_{i})-f(t_{i-1})}{t_{i}-t_{i-1}}=\frac{\Delta{x}_{i}}{\Delta{t}}=f'(t_{i}^{*})\\
        g_{i}'=\frac{g(t_{i})-g(t_{i-1})}{t_{i}-t_{i-1}}=\frac{\Delta{y}_{i}}{\Delta{t}}=g'(t_{i}^{**})
        \end{align*}
        $$
        <figcaption style='font-size:15px'> 
            Sejam $t_{i}^{*}$ e $t_{i}^{**}$ valores em $[t_{i-1},t_{i}]$ cuja derivada equivale à inclinação $\frac{f(t_{i})-f(t_{i-1})}{t_{i}-t_{i-1}}$ e $\frac{g(t_{i})-g(t_{i-1})}{t_{i}-t_{i-1}}$, seguindo o Mean Value Theorem.
        </figcaption>
    </center>
</figure>

<div> 
    <ul style='font-size:20px'>
        <li>
            A partir desses valores, poderemos definir $\overline{P_{i-1}P_{i}}$ e $L$ da seguinte forma:
        </li>
     </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'> 
        $$
        \displaystyle
        \begin{align*}
            \overline{P_{i-1}P_{i}}&=\sqrt{(f'(t_{i}^{*})\Delta{t})^{2}+(g'(t_{i}^{**})\Delta{t})^{2}}\\
            &=\sqrt{f'(t_{i}^{*})^{2}+g'(t_{i}^{**})^{2}}\Delta{t}\\
            L&=\lim_{n\to\infty}{\sum_{i=1}^{n}{\sqrt{f'(t_{i}^{*})^{2}+g'(t_{i}^{**})^{2}}\Delta{t}}}
        \end{align*}
        $$
        <figcaption style='font-size:15px'> 
        </figcaption>
    </center>
</figure>

<div> 
    <ul style='font-size:20px'>
        <li>
            Repare que agora poderemos expressar $L$ como uma Soma de Riemann e, com isso, defini-lo como uma integral!
        </li>
        <li>
            <strong>Nota:</strong> Por mais que $t_{i}^{*}$ e $t_{i}^{**}$ não necessariamente serão o mesmo número, podemos assumir que o são por conta da natureza infinitesimal de $[t_{i-1},t_{i}]$.
        </li>
     </ul>
</div>

<figure>
    <center style='font-size:20px;margin-top:20px'> 
        $$
        \displaystyle
        L=\int_{\alpha}^{\beta}{\sqrt{\left(\frac{dx}{dt}\right)^{2}+\left(\frac{dy}{dt}\right)^{2}}dt}
        $$
        <figcaption style='font-size:15px'> 
        </figcaption>
    </center>
</figure>

<div> 
    <ul style='font-size:20px'>
        <li>
            É importante frisar que essa fórmula é válida para apenas curvas que passam uma única vez sobre o seu traçado. Caso elas passem $n$ vezes, seu valor retornado será $nL$. 
        </li>
     </ul>
</div>

<h2 style='font-size:30px'> 10.3 Polar Coordinates</h2>
<div> 
    <ul style='font-size:20px'> 
        <li>
            Em matemática, os sistemas de coordenadas são formas de representarmos pontos presentes num plano por um par ordenado de números. O principal sistema ensinado seria o Cartesiano, em que as coordenadas seriam as distâncias do ponto a cada eixo.
        </li>
        <li>
            Nessa seção, estudaremos um outro conhecido como Polar Coordinate System, que pode ser mais conveniente para certos casos. 
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> O Sistema de Coordenadas Polares</h3>
<div> 
    <ul style='font-size:20px'>
        <li>
            O sistema se inicia com o que chamamos de <i>pole</i>, equivalente à origem no Cartesiano. A partir dele, traçamos um raio à sua direita, correspondendo ao eixo-x do Plano Cartesiano, sendo esse chamado de <i> polar axis</i>.
        </li>
        <li>
            Todo ponto $P$ presente nesse plano com distância $r$ do polo, em que a semi-reta $\overline{OP}$ forme um ângulo $\theta$ com o eixo polar, é representado como $(r,\theta)$. No caso, $r$ e $\theta$ seriam reconhecidos como <i> polar coordinates</i> de $P$.
        </li>
        <li>
            Nesse sistema, um ângulo será positivo caso seja medido no sentido anti-horário. Caso o contrário, será negativo. Além disso, repare que sempre que $r=0$, estaremos nos referindo ao pole.
        </li>
        <li>
            Admitimos que $r$ assuma qualquer valor real. Caso $r<0$, o ponto estará localizado a uma distância $|r|$ do polo, mas no quadrante contrário em que $\theta$ estiver representado. Se $r\geq{0}$, o ponto estará no mesmo quadrante que $\theta$.
        </li>
     </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Symmetry</h3>
