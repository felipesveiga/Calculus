<h2 style='font-size:30px'> 4.1 Maximum and Minimum Values</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Um valor f(c) é considerado máximo global da função f(x) caso ele não seja menor do que qualquer outro número do conjunto Imagem; o oposto é válido para o mínimo global.
        </li>
        <li> 
            O máximo ou mínimo locais serão os valores mais altos ou baixos da função dentro de um intervalo restrito.
        </li>
    </ul>
</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# Uma função pode ter apenas um mínimo global, e não um máximo global.
x = np.linspace(-10, 10, 1001)
plt.plot(x, x**2)
plt.title('x²')
```

```python
# Os máximos e mínimos globais podem se repetir infinitamente.
plt.plot(x, np.sin(x))
plt.axhline(1, color='k',ls='--', lw=.75)
plt.axhline(-1, color='k',ls='--', lw=.75)

plt.title('sin(x)')
```

```python
# Ou ainda, os máximos e mínimos globais podem sequer existir!
plt.plot(x, x**3)
plt.title('x³')
```

<h3 style='font-size:30px;font-style:italic'> Extreme Value Theorem</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            O Teorema do Valor Extremo  define os requisitos que uma função deve possuir para que ela tenha máximos e mínimos globais. Esses seriam:
            <ul style='list-style-type:lower-alpha'> 
                <li> 
                    A função deve ser contínua.
                </li>
                <li> 
                    Deve conter um intervalo fechado.
                </li>
            </ul>
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Fermat's Theorem</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Até o momento, identificamos como uma função pode ter valores extremos. Mas como é possível saber quais são eles?
        </li>
        <li> 
            O Teorema de Fermat oferece uma explicação bastante simples. Se f(c) é o máximo ou mínimo global, f'(c)=0.  
        </li>
        <li> 
            Não seria possível f'(c)>0, se f(c) é o valor máximo; é impossível f'(c) &lt;0, se f(c) for o mínimo global.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Cuidados com o Teorema de Fermat</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Observe que não podemos radicalizar o Teorema de Fermat. Nem toda abscissa c em que f'(c)=0 é aquela em que a função recebe os seu valores extremos. Às vezes, a derivada da abscissa dos pontos máximos ou mínimos pode sequer existir. 
        </li>
    </ul>
</div>

```python
# Por exemplo, mesmo que a derivada de f(x)=x³ em 0 seja igual a 0, esse ponto não corresponde ao mínimo ou máximo da função!
x = np.linspace(-10, 10, 1001)
plt.plot(x, x**3)
plt.scatter(0,0, color='k')
```

```python
# A |x|, mesmo tendo o seu mínimo global em x=0, sua derivada nessa abscissa não existe!
from sympy import Abs, symbols, diff
x = symbols('x')
diff(Abs(x)).subs(x,0)
```

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Critical Number</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Os números críticos de f(x) são aqueles em que a derivada da função é 0, ou em que essa mesma não existe. 
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example VII</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Descubra os números críticos de $f(x)=x^{\frac{3}{5}}(4-x)$
        </li>
    </ul>
</div>

```python
from sympy import symbols, diff, Eq, solve
x = symbols('x')
d = diff(x**(3/5) * (4 - x))

# Para quais valores de x a derivada é 0?
print(solve(Eq(d,0)))

# Observe que, como o divisor da equação é x, se x=0, receberemos NaN como resultado.
d

# Portanto, os pontos críticos da função sejam (3/2, 0)
```

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> The Closed Interval Method</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Para encontrar os valores extremos de uma função num intervalo fechado [a,b], tente seguir esses três passos:
            <ul style='list-style-type:lower-alpha'> 
                <li> 
                    Encontrar os números críticos.
                </li>
                <li> 
                    Medir o valor da função nos pontos a,b.
                </li>
                <li> 
                    Escolher as abscissas que retornem o maior e menor valor para a função.
                </li>
            </ul>
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example VIII</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Descubra os valores máximos e mínimos de $f(x)=x^{3}-3x^{2}+1$ em $-\frac{1}{2}\leq x \leq 4$
        </li>
    </ul>
</div>

```python
from sympy import diff, Eq, solve
from sympy.abc import x
# Diferenciando a equação.
fx = x**3 - 3*x**2 + 1
dx = diff(fx)

# Obtendo os valores críticos.
criticals = solve(Eq(dx,0))
print(f'Criticals:{criticals}')

# Será que os pontos extremos do intervalo possuem os valores extremos?
f_12, f4 = fx.subs(x, -1/2), fx.subs(x,4)

# Podemos analisar que os pontos máximo e mínimo ocorrerão quando x=4 e x=2, respectivamente.
f_12, f4, [fx.subs(x,i) for i in criticals]
```

```python
# O sympy possui algumas classes prontas para fazermos esse tipo de avaliação:
from sympy import maximum, minimum,Interval
maximum(fx, x, Interval(-1/2, 4)), minimum(fx, x, Interval(-1/2, 4))
```

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example IX</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Descubra os valores máximos e mínimos de $f(x)=x - 2\sin{x}$ em $0\leq x \leq 2\pi$
        </li>
    </ul>
</div>

```python
from sympy import diff, sin, Eq, solve, pi
from sympy.abc import x

# Igualando a diferenciação a zero.
fx = x - 2*sin(x)
dx = diff(fx)
criticals = solve(Eq(dx,0))

# Acrescentando a 'criticals' os valores-limite do intervalo.
criticals.extend([0, 2*pi])

# Medindo f(x) para cada valor de 'criticals'.
d = {xi:fx.subs(x,xi) for xi in criticals}

# Qual o número mínimo e máximo da função?
min(d.values()), max(d.values())
```

```python
# Novamente, fazendo a correção do exercício pelo sympy.
from sympy import maximum, minimum, Interval
minimum(fx, x, Interval(0, 2*pi)),maximum(fx, x, Interval(0, 2*pi))
```

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example X</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Sendo a velocidade do telecópio Hubble entre sua decolagem (t=0s) e a ativação de suas turbinas (t=126s) $v(t)=0.001302t^{3}-0.09029t^{2}+23.61t-3.083$, descubra o valor mínimo e máximo da aceleração.
        </li>
    </ul>
</div>

```python
# A função da aceleração deve ser descoberta diferenciando a v(t).
from sympy import diff, sin, Eq, solve, pi
from sympy.abc import t
vt = 0.001302*t**3 - 0.09029*t**2+23.61*t - 3.083
at = diff(vt)

# Obtendo os valores críticos de a(t).
jt = diff(at)
criticals = solve(Eq(jt,0))
criticals.extend([0, 126])

# Medindo a a(t) para cada número em 'criticals'.
d = {xi:at.subs(t,xi) for xi in criticals}

# Desvendando os números extremos de a(t).
min(d.values()), max(d.values())
```

```python
# Correção:
from sympy import maximum, minimum, Interval
minimum(at, t, Interval(0, 126)), maximum(at, t, Interval(0,126))
```

<h2 style='font-size:30px'> 4.2 The Mean Value Theorem</h2>

<h3 style='font-size:30px;font-style:italic'> The Rolle's Theorem</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            O Teorema de Rolle afirma que uma função f(x) terá uma derivada f'(c)=0 em alguma abscissa c de seu domínio caso as seguintes condições sejam satisfeitas:
            <ul style='list-style-type:lower-alpha'> 
                <li> 
                    A função seja contínua no intervalo [a,b].
                </li>
                <li> 
                    f seja diferenciável no intervalo <strong> aberto</strong> (a,b).
                </li>
                <li> 
                    f(a)=f(b)
                </li>
            </ul>
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> The Mean Value Theorem</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            O Teorema do Valor Médio postula que uma função f que respeita as seguintes condições terá um ponto c cuja tangente terá a mesma inclinação do que a reta que liga diretamente os pontos a e b.
        </li>
            <ul style='list-style-type:lower-alpha'> 
                <li> 
                    A função seja contínua no intervalo [a,b].
                </li>
                <li> 
                    f seja diferenciável no intervalo <strong> aberto</strong> (a,b).
                </li>
                $$
                    f'(c)=\frac{f(b)-f(a)}{b-a}
                $$
        </ul> 
        <li> 
            Uma outra interpretação, mais condizente com o nome do teorema, é de que um ponto c terá a sua tangente com a inclinação igual à taxa de variação média da função no intervalo [a,b].
        </li>
    </ul>
</div>

```python
# Observe o caso de f(x)=x² no intervalo [-5,5].
x = np.linspace(-5, 5, 101)
plt.plot(x, x**2)
plt.axhline(25, ls='--', color='k')
plt.axhline(0, ls='--', color='k')
```
