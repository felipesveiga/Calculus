<h2 style='font-size:30px'> 4.1 Maximum and Minimum Values</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Um valor f(c) é considerado máximo global da função f(x) caso ele não seja menor do que qualquer outro número do conjunto Imagem; o oposto é válido para o mínimo global.
        </li>
        <li> 
            O máximo ou mínimo locais serão os valores mais altos ou baixos da função dentro de um intervalo restrito.
        </li>
    </ul>
</div>

```python
import matplotlib.pyplot as plt
import numpy as np

# Uma função pode ter apenas um mínimo global, e não um máximo global.
x = np.linspace(-10, 10, 1001)
plt.plot(x, x**2)
plt.title('x²')
```

```python
# Os máximos e mínimos globais podem se repetir infinitamente.
plt.plot(x, np.sin(x))
plt.axhline(1, color='k',ls='--', lw=.75)
plt.axhline(-1, color='k',ls='--', lw=.75)

plt.title('sin(x)')
```

```python
# Ou ainda, os máximos e mínimos globais podem sequer existir!
plt.plot(x, x**3)
plt.title('x³')
```

<h3 style='font-size:30px;font-style:italic'> Extreme Value Theorem</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            O Teorema do Valor Extremo  define os requisitos que uma função deve possuir para que ela tenha máximos e mínimos globais. Esses seriam:
            <ul style='list-style-type:lower-alpha'> 
                <li> 
                    A função deve ser contínua.
                </li>
                <li> 
                    Deve conter um intervalo fechado.
                </li>
            </ul>
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Fermat's Theorem</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Até o momento, identificamos como uma função pode ter valores extremos. Mas como é possível saber quais são eles?
        </li>
        <li> 
            O Teorema de Fermat oferece uma explicação bastante simples. Se f(c) é o máximo ou mínimo global, f'(c)=0.  
        </li>
        <li> 
            Não seria possível f'(c)>0, se f(c) é o valor máximo; é impossível f'(c) &lt;0, se f(c) for o mínimo global.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Cuidados com o Teorema de Fermat</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Observe que não podemos radicalizar o Teorema de Fermat. Nem toda abscissa c em que f'(c)=0 é aquela em que a função recebe os seu valores extremos. Às vezes, a derivada da abscissa dos pontos máximos ou mínimos pode sequer existir. 
        </li>
    </ul>
</div>

```python
# Por exemplo, mesmo que a derivada de f(x)=x³ em 0 seja igual a 0, esse ponto não corresponde ao mínimo ou máximo da função!
x = np.linspace(-10, 10, 1001)
plt.plot(x, x**3)
plt.scatter(0,0, color='k')
```

```python
# A |x|, mesmo tendo o seu mínimo global em x=0, sua derivada nessa abscissa não existe!
from sympy import Abs, symbols, diff
x = symbols('x')
diff(Abs(x)).subs(x,0)
```

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> The Closed Interval Method</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Para encontrar os valores extremos de uma função num intervalo fechado [a,b], tente seguir esses três passos:
            <ul style='list-style-type:lower-alpha'> 
                <li> 
                    Encontrar os números críticos.
                </li>
                <li> 
                    Medir o valor da função nos pontos a,b.
                </li>
                <li> 
                    Escolher as abscissas que retornem o maior e menor valor para a função.
                </li>
            </ul>
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example VIII</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Descubra os valores máximos e mínimos de $f(x)=x^{3}-3x^{2}+1$ em $-\frac{1}{2}\leq x \leq 4$
        </li>
    </ul>
</div>

```python
from sympy import diff, Eq, solve
from sympy.abc import x
# Diferenciando a equação.
fx = x**3 - 3*x**2 + 1
dx = diff(fx)

# Obtendo os valores críticos.
criticals = solve(Eq(dx,0))
print(f'Criticals:{criticals}')

# Será que os pontos extremos do intervalo possuem os valores extremos?
f_12, f4 = fx.subs(x, -1/2), fx.subs(x,4)

# Podemos analisar que os pontos máximo e mínimo ocorrerão quando x=4 e x=2, respectivamente.
f_12, f4, [fx.subs(x,i) for i in criticals]
```

```python
# O sympy possui algumas classes prontas para fazermos esse tipo de avaliação:
from sympy import maximum, minimum,Interval
maximum(fx, x, Interval(-1/2, 4)), minimum(fx, x, Interval(-1/2, 4))
```

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example IX</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Descubra os valores máximos e mínimos de $f(x)=x - 2\sin{x}$ em $0\leq x \leq 2\pi$
        </li>
    </ul>
</div>

```python
from sympy import diff, sin, Eq, solve, pi
from sympy.abc import x

# Igualando a diferenciação a zero.
fx = x - 2*sin(x)
dx = diff(fx)
criticals = solve(Eq(dx,0))

# Acrescentando a 'criticals' os valores-limite do intervalo.
criticals.extend([0, 2*pi])

# Medindo f(x) para cada valor de 'criticals'.
d = {xi:fx.subs(x,xi) for xi in criticals}

# Qual o número mínimo e máximo da função?
min(d.values()), max(d.values())
```

```python
# Novamente, fazendo a correção do exercício pelo sympy.
from sympy import maximum, minimum, Interval
minimum(fx, x, Interval(0, 2*pi)),maximum(fx, x, Interval(0, 2*pi))
```

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example X</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Sendo a velocidade do telecópio Hubble entre sua decolagem (t=0s) e a ativação de suas turbinas (t=126s) $v(t)=0.001302t^{3}-0.09029t^{2}+23.61t-3.083$, descubra o valor mínimo e máximo da aceleração.
        </li>
    </ul>

```python
# A função da aceleração deve ser descoberta diferenciando a v(t).
from sympy import diff, sin, Eq, solve, pi
from sympy.abc import t
vt = 0.001302*t**3 - 0.09029*t**2+23.61*t - 3.083
at = diff(vt)

# Obtendo os valores críticos de a(t).
jt = diff(at)
criticals = solve(Eq(jt,0))
criticals.extend([0, 126])

# Medindo a a(t) para cada número em 'criticals'.
d = {xi:at.subs(t,xi) for xi in criticals}

# Desvendando os números extremos de a(t).
min(d.values()), max(d.values())
```

<h2 style='font-size:30px'> 4.2 The Mean Value Theorem</h2>

<h3 style='font-size:30px;font-style:italic'> The Rolle's Theorem</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            O Teorema de Rolle afirma que uma função f(x) terá uma derivada f'(c)=0 em alguma abscissa c de seu domínio caso as seguintes condições sejam satisfeitas:
            <ul style='list-style-type:lower-alpha'> 
                <li> 
                    A função seja contínua no intervalo [a,b].
                </li>
                <li> 
                    f seja diferenciável no intervalo <strong> aberto</strong> (a,b).
                </li>
                <li> 
                    f(a)=f(b)
                </li>
            </ul>
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> The Mean Value Theorem</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            O Teorema do Valor Médio postula que uma função f que respeita as seguintes condições terá um ponto c cuja tangente terá a mesma inclinação do que a reta que liga diretamente os pontos a e b.
        </li>
            <ul style='list-style-type:lower-alpha'> 
                <li> 
                    A função seja contínua no intervalo [a,b].
                </li>
                <li> 
                    f seja diferenciável no intervalo <strong> aberto</strong> (a,b).
                </li>
                $$
                    f'(c)=\frac{f(b)-f(a)}{b-a}
                $$
        </ul> 
        <li> 
            Uma outra interpretação, mais condizente com o nome do teorema, é de que um ponto c terá a sua tangente com a inclinação igual à taxa de variação média da função no intervalo [a,b].
        </li>
    </ul>
</div>

```python
# Observe o caso de f(x)=x² no intervalo [-5,5].
x = np.linspace(-5, 5, 101)
plt.plot(x, x**2)
plt.axhline(25, ls='--', color='k')
plt.axhline(0, ls='--', color='k')
```

<div> 
    <ul style='font-size:20px'> 
        <li> 
            Nesta seção, somos apresentados a uma consequência mais óbvia das derivadas. Se f'(c)>0, a função está crescendo em c; o caso contrário indica que ela decresce.

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example I</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Descubra onde $f(x)=3x^{4}-4x^{3}-12x^{2}+5$ aumenta ou diminui.
        </li>
    </ul>
</div>

```python
# O procedimento parece ser simples. Vamos extrair as raízes de f'(x) e saber o valor dessa mesma nos pontos anteriores às abscissas 
# coletadas.
from sympy import diff, Eq, solve, limit, oo
from sympy.abc import x

fx = 3*x**4 - 4*x**3 - 12*x**2 + 5

# Derivando f(x).
dydx = diff(fx)

# Obtendo as raízes de dy/dx.
a, b, c = solve(Eq(dydx,0))

print(f'Raízes: {a,b,c}')

# A função decresce em [-oo, -1)
print(limit(dydx, x, -oo))

# A função sobe em [-1, 0]
print(limit(dydx, x, (a+b)/2))

# A função decresce em [0, 2].
print(limit(dydx, x, (0+2)/2))

# A função sobe em [2, +oo]
print(limit(dydx, x, oo))
```

```python
# Experimente o 'is_increasing' ou 'is_decreasing', do sympy!
from sympy import is_increasing, Interval
is_increasing(fx, Interval(-oo, -1))
```

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> The First Derivative Test</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
             O Teste da Primeira Derivada analisa os valores críticos de uma função e avalia se esses consistem em pontos de inflexão (máximos ou mínimos). Para um ponto crítico c:
            <ul style='list-style-type:lower-alpha'> 
                <li> 
                    Se f' deixa de ser positiva para se tornar negativa em c, há um máximo local em c.
                </li>
                <li> 
                    Se f' deixa de ser negativa para se tornar positiva em c, há um mínimo local em c.
                </li>
                <li> 
                    Se não há alteração no sinal de f', não existe máximo ou mínimo em c.
                </li>
            </ul>
        </li>
    </ul>
</div>

```python
# Essa regra justifica o fato de o ponto (0,0) não ser um mínimo ou máximo para f(x)=x³.
# A derivada da função não se torna negativa nesse local.
x = np.linspace(-10, 10, 101)
plt.plot(x, x**3)
```

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example II</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Encontre os valores máximo e mínimo da função do Exemplo I.
        </li>
    </ul>
</div>

```python
from sympy.abc import x
from sympy import diff, Eq, solve
fx = 3*x**4 - 4*x**3 - 12*x**2 + 5

# Derivando f(x).
dydx = diff(fx)

minimum = float('inf')
maximum = -float('inf')

# Como a função não possui C.E, apenas a solução de quando dydx=0 são os números críticos.
criticals = solve(Eq(dydx,0))

for number in criticals:
    fnumber = fx.subs(x,number)
    if fnumber<minimum:
        minimum = fnumber
    elif fnumber>maximum:
        maximum = fnumber
    else:
        continue

print(f'Minimum: {minimum}; Maximum: {maximum}')
```

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example III</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Descubra os valores máximo e mínimo de $g(x)=x + 2\sin{x}$ em $0 \leq x \leq 2\pi$. 
        </li>
    </ul>
</div>

```python
from sympy import sin, diff, Eq, solve, maximum, minimum, Interval, pi
from sympy.abc import x

# Diferenciando a função.
gx = x + 2 * sin(x)
dydx = diff(gx)

# Mesmo caso do Exemplo II: não há C.E para a função. Só temos que estar atentos se os números de 'criticals' estão contidos no intervalo.
criticals = solve(Eq(dydx, 0))

maximum = -float('inf')
minimum = float('inf')

for number in criticals:
    fnumber = gx.subs(x, number)
    if fnumber < minimum:
        minimum = fnumber
    if fnumber > maximum:
        maximum = fnumber
    else:
        continue
        
print(f'Maximum: {maximum}; Minimum: {minimum}')
```

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example VI</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Considerando o Teste da Segunda Derivada, encontre os valores máximo e mínimo de $f(x)=x^{4}-4x^{3}$. 
        </li>
    </ul>
</div>

```python
from sympy import diff, Eq, solve
from sympy.abc import x
fx = x**4 - 4*x**3

# Encontrando a primeira derivada da função.
dydx = diff(fx,x,1)

# A função não tem C.E, então os valores de dydx=0 bastam.
criticals = solve(Eq(dydx, 0))

dy2dx = diff(fx,x,2)

for critical in criticals:
    sec_der = dy2dx.subs(x, critical)
    if sec_der < 0:
        print(f'Maximum: {(critical, fx.subs(x,critical))}')
    elif sec_der > 0:
        print(f'Minimum: {(critical, fx.subs(x,critical))}')
    else:
        continue
```

<h2 style='font-size:30px'> 4.4 Indeterminate Forms and L'Hospital Rule</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Existem casos de limites cujos resultados podem adquirir uma forma indeterminada. Veja a equação abaixo:
            <p style='margin-top:10px'>
            $$
                \lim_{x \to 1} \frac{\ln{x}}{x-1}
            $$
            </p>
            <ul style='list-style-type:circle'> 
                <li> 
                    O resultado que obteremos apenas por imputar o 1 na equação é $\frac{0}{0}$, um valor não definido. Vimos em outros capítulos que existem técnicas que tentam contornar esse problema, seja por fatoração, seja dividindo as partes da fração pela maior potência de x do denominador. Mas nenhuma dessas estratégias é possível nesse caso.
                </li>
                <li> 
                    Portanto, estamos diante da <strong>forma indefinida do tipo 0/0</strong>.
                </li>
            </ul>
        </li>
        <li> 
            Além disso, há situações em que o limite retornado tenderá a $\frac{\pm \infty}{\pm\infty}$. A função $f(x)=\frac{\ln{x}}{x-1}$
passará por esse problema quando $x \to \infty$.
            <ul style='list-style-type:circle'> 
                <li> 
                    Nesse caso, existe uma <strong> forma indefinida do tipo $\infty/\infty$</strong>.
                </li>
            </ul>
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> L'Hospital Rule</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Para aplicar essa regra é necessário ter duas funções, f(x) e g(x), diferenciáveis em um intervalo que contém a, com $g'(x) \neq 0$. Caso $\lim_{x \to a}\frac{f(x)}{g(x)}$ apresente uma das formas indefinidas apresentadas (0/0 ou $\infty/\infty$) por mera substituição de valores: 
            $$\lim_{x \to a}\frac{f(x)}{g(x)}=\lim_{x \to a}\frac{f'(x)}{g'(x)}$$.
        </li>
        <li> 
            Nota: A Regra de L'Hospital é aplicável também a limites unilaterais!
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example I</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Encontre  $\lim_{x \to 1}\frac{\ln{x}}{x-1}$
        </li>
    </ul>
</div>

```python
# O sympy é capaz de aplicar a Regra de L'Hospital por conta própria, mas vamos descobrir o limite de maneira mais manual neste exemplo.
from sympy.abc import x
from sympy import diff, Eq, solve, log

ax = log(x)
bx = x - 1

# Encontrando o limite do quociente entre as diferenciações.
(diff(ax) / diff(bx)).subs(x,1)
```

<h3 style='font-size:30px;font-style:italic'> Indeterminate Products</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Se $\lim_{x \to a}f(x)=0$ e $\lim_{x \to a}g(x)=\infty$,, o limite $\lim_{x \to a}f(x)g(x)$ é indeterminado (forma $0 \cdot \infty$). O motivo de essa indeterminação existir fica mais evidente se escrevermos fg das seguintes maneiras:
            <p style='margin-left:400px'> 
                $fg=\frac{f}{\frac{1}{g}}$ ou $fg=\frac{g}{\frac{1}{f}}$
            </p>
        </li>
        <li> 
            Com as transformações apresentadas, nos induzimos a cair nas formas indeterminadas $0/0$ ou $\infty/\infty$. 
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example VI</h4>

<h3 style='font-size:30px;font-style:italic'> Indeterminate Powers</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Quando $\lim_{x \to a}[f(x)]^{g(x)}$ nos dá como resultado $0^{0}$, $\infty^{0}$ ou $1^{\infty}$, existe uma indeterminação de potência.
        </li>
        <li> 
            Para nos livrarmos dessa situação, o recomendável é extrairmos o logaritmo natural da expressão e, logo em seguida, transformarmos esse em uma função exponencial: 
            $$
                \ln{y}=g(x)\ln{f(x)} \rightarrow y=e^{g(x)\ln{f(x)}}
            $$
        </li>
        <li> 
            O limite de $\ln{y}$ poderá nos induzir a cair na indeterminação $\infty \cdot 0$, obrigando-nos a usar L'Hospital. Uma vez medido esse limite, basta usá-lo como expoente do número de Euler.
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example VIII</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Encontre  $\lim_{x \to 0^{+}}(1 + \sin{4x})^{\cot{x}}$
        </li>
    </ul>
</div>

```python
# Por mera substituição, teremos como resultado a indeterminação 1**oo.
# Vamos extrair o logaritmo natural da função.

from sympy import sin, cot, diff, limit, log, tan, exp
from sympy.abc import x

lny = log((1 + sin(4*x)) ** (1/tan(x)))

# Observe que se tombarmos (1/tan(x)), cairemos na indeterminação 0/0.
lny.doit()
```

```python
# Para solucionar isso, usaremos L'Hospital.
dlnydx = diff(log(sin(4*x)+1)) / diff(tan(x))
dlnydx
```

```python
# Agora, basta uasr o limite de dlnydx para 0⁺ como expoente de e. Esse será o limite de y.
exp(dlnydx.subs(x, 0))
```

<h2 style='font-size:30px'> 4.8 Newton's Method</h2>
<div> 
    <ul style='font-size:20px'> 
        <li> 
             O Método de Newton é um algoritmo que busca por raízes de funções. Precisamos, em primeiro lugar, de um ponto qualquer na função de onde a derivaremos a fim de encontrarmos a sua linearização (ver seção 3.10).
        </li>
        <li> 
            A aproximação linear da função será o x em que a equação da tangente é 0. Se quisermos ter um valor mais preciso, conseguimos, agora, aproximarmos a função com a reta tangente nesse mesmo x. Esse processo pode ser repetido quantas vezes quisermos!
        </li>
        <li> 
            É importante salientar que nem sempre a aproximação convergirá ao resultado real, mas caso o faça, a notação é a seguinte (n é a quantidade de iterações; r é a raiz da função):
            $$
                \lim_{n \to \infty}x_n=r
            $$
        </li>
    </ul>
</div>

<h3 style='font-size:30px;font-style:italic'> Precisão do método de Newton</h3>
<div> 
    <ul style='font-size:20px'> 
        <li> 
             Se quisermos que a solução encontrada pelo método seja correta nas p-primeiras casas decimais, uma coisa que podemos fazer é comparar $x_n$ e $x_{n+1}$ e verificar se essas casas são c
        </li>
    </ul>
</div>

<h4 style='font-size:30px;font-style:italic;text-decoration:underline'> Example II</h4>
<div> 
    <ul style='font-size:20px'> 
        <li> 
            Encontre a aproximação de $\lim_{x \to 0 ^{+}}x^{x}$ com as 8 primeiras casas decimais iguais às da solução real.
        </li>
    </ul>
</div>

```python
from sympy import Eq, solve, diff, cos
from sympy.abc import x
import numpy as np
class NewtonMethod():
    
    def __init__(self, ixt:float, eq:Eq, correct_decimals=5, n_iter=None):
        # Ponto de tangência inicial.
        self.ixt = ixt
        # Equação da função.
        self.eq = eq
        # Quantidade de casas decimais que devem estar corretas.
        self.correct_decimals = correct_decimals
        # Número de iterações para o encontro da solução.
        self.n_iter = n_iter

        assert (isinstance(self.correct_decimals, int) and self.n_iter is None) or (self.correct_decimals is None and 
                                                                                      isinstance(self.n_iter, int))
        
    def __post_init__(self):
        # '__diff' é a primeira derivada de 'eq'.
        self.__diff = diff(self.eq,x)
        # '__xt' é o ponto de tangência da última iteração.
        self.__xt = self.ixt
        
    def find_root(self):
        self.__post_init__()
        # '__lxt' armazena todos os pontos de tangência calculados.
        self.__lxt = np.array([0, np.inf])
        if self.correct_decimals is not None:
            while abs(self.__lxt[-2] - self.__lxt[-1])>10**-self.correct_decimals:
                self.__xt -= float((self.eq.subs({'x':self.__xt}) / self.__diff.subs({'x':self.__xt})))
                self.__lxt = np.append(self.__lxt, self.__xt)
            return self.__xt
        else:
            for i in range(self.n_iter):
                self.__xt -= float((self.eq.subs({'x':self.__xt}) / self.__diff.subs({'x':self.__xt})))
            return self.__xt

a = NewtonMethod(1, -x**2 +3*x+27, 10)
b = a.find_root()
b
```
